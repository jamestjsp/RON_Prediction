```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
columns = ['Sample_#', 'Reformer_WAIT', 'Feed_Mass_Flow', 'H2HC_Ratio', 'C5+_RON', 'C6+_RON', 
           'Total_aromatics', 'Total_C8_aromatics', 'Wt_Percentage_A6', 'Wt_Percentage_A7', 
           'Wt_Percentage_H2', 'Wt_Percentage_P1', 'Wt_Percentage_P2', 'Wt_Percentage_P3', 
           'Wt_Percentage_O3', 'Wt_Percentage_ETHYLBEN', 'Wt_Percentage_M-XYLENE', 
           'Wt_Percentage_O-XYLENE', 'Wt_Percentage_P-XYLENE', 'Reformate_RON', 'Coke_Laydown_Rate_Rx1', 
           'Coke_Laydown_Rate_Rx2', 'Coke_Laydown_Rate_Rx3', 'Coke_Laydown_Rate_Rx4']
```

```python
CatRef_Data = pd.read_excel("Case_study_20190812.xlsx", sheet_name="Case_study_20190812",names = columns,index_col=0)
CatRef_Data.head()
```

```python
# CatRef_Data.columns
data_HbyHC3_0 = CatRef_Data.loc[(CatRef_Data['Feed_Mass_Flow'] == 170000) & (CatRef_Data['H2HC_Ratio'] == 3.0)]
data_HbyHC3_5 = CatRef_Data.loc[(CatRef_Data['Feed_Mass_Flow'] == 170000) & (CatRef_Data['H2HC_Ratio'] == 3.5)]
data_HbyHC4_0 = CatRef_Data.loc[(CatRef_Data['Feed_Mass_Flow'] == 170000) & (CatRef_Data['H2HC_Ratio'] == 4.0)]
```

```python
fig = plt.figure()

for frame in [data_HbyHC3_0, data_HbyHC3_5, data_HbyHC4_0]:
    plt.plot(frame['Reformer_WAIT'], frame['C6+_RON'], )
   

fig.legend(('H2HC=3.0', 'H2HC=3.5', 'H2HC=4.0'), bbox_to_anchor=(.85,0.35))
plt.xlim(485,540)
plt.ylim(90,108)
plt.show()
```

```python
fig = plt.figure()

for frame in [data_HbyHC3_0, data_HbyHC3_5, data_HbyHC4_0]:
    plt.plot(frame['Reformer_WAIT'], frame['Coke_Laydown_Rate_Rx4'])

fig.legend(('H2HC=3.0', 'H2HC=3.5', 'H2HC=0.0'), bbox_to_anchor=(.85,0.35))
plt.xlim(485,540)
plt.ylim(0,145)
plt.show()
```

```python
fig = plt.figure()

for frame in [data_HbyHC3_0, data_HbyHC3_5, data_HbyHC4_0]:
    plt.plot(frame['Reformer_WAIT'], frame['Wt_Percentage_A6'])

fig.legend(('H2HC=3.0', 'H2HC=3.5', 'H2HC=0.0'), bbox_to_anchor=(.8,0.35))
plt.xlim(485,540)
plt.ylim(2,7)
plt.show()
```
